```python
import torch
from torch.utils.data import TensorDataset, DataLoader
```

```python
n_train = 406
n_lines = 506
n_cols = 13
data = torch.tensor(())
data = data.new_zeros((n_lines,n_cols))
labels = torch.tensor(())
labels = labels.new_zeros((n_lines))
f = open('housing.csv')
n_line = 0
for line in f:
    cols = line.split()
    for i in range(n_cols-1):
        data[n_line][i] = float(cols[i])
    labels[n_line] = float(cols[n_cols])
    n_line += 1
    
f.close()
```

```python
# варим фичи :)
n_train_lines = 400
data_norm = torch.tensor(())
data_norm = data.new_zeros((n_lines,n_cols))
data_min_max = torch.tensor(())
data_min_max = data.new_zeros((n_cols, 3))

for i in range(n_cols):
    data_min_max[i][0] = 1000000  # set min
    data_min_max[i][1] = -1000000   # set max for our simple dataset

for i in range(n_lines):
    for j in range(n_cols):
        if data[i][j] < data_min_max[j][0]:
            data_min_max[j][0] = data[i][j]
        if data[i][j] > data_min_max[j][1]:
            data_min_max[j][1] = data[i][j]    

for i in range(n_cols):
    # calculate disperse
    data_min_max[i][2] = data_min_max[i][1] - data_min_max[i][0]
    
# nomralize data
for i in range(n_lines):
    for j in range(n_cols):
        data_norm[i][j] = (data[i][j] - data_min_max[j][0]) / data_min_max[j][2]

data_train = torch.tensor(())
data_train = data.new_zeros((n_train_lines,n_cols))
labels = torch.tensor(())
labels = labels.new_zeros((n_train_lines))

data_train = data_norm[:n_train_lines]
labels_train = labels[:n_train_lines]

```

```python
batch_size = 10
dataset = TensorDataset(data_train, labels_train)
data_iter = DataLoader(dataset, batch_size, shuffle=True)
```

```python
model = torch.nn.Sequential(torch.nn.Linear(n_cols, 1))
loss = torch.nn.MSELoss(reduction='mean')
trainer = torch.optim.SGD(model.parameters(), lr=0.00001)
```

```python
num_epochs = 20
for epoch in range(0, num_epochs):
    for X, y in data_iter:
        trainer.zero_grad()
        #print (X,y)
        l = loss(model.forward(X).reshape(-1), y)
        print (l)
        l.backward()
        trainer.step()
    l = loss(model.forward(data_train).reshape(-1), labels_train)
    print('epoch %d, loss: %f' % (epoch, l.item()))
    print('w', model[0].weight.data)
    print('b', model[0].bias.data)
```
